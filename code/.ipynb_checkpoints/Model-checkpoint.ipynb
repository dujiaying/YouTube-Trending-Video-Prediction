```python
# load the packages
import json  
import re
from collections import Counter
import time
import numpy as np
import pandas as pd
from pandas.io.json import json_normalize  
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from imblearn.pipeline import Pipeline 
from imblearn.over_sampling import ADASYN

from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import RandomizedSearchCV

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import SGDClassifier
from sklearn.ensemble import GradientBoostingClassifier

from sklearn.metrics import roc_auc_score, accuracy_score, average_precision_score, classification_report, precision_score

import shap
import lime

%matplotlib inline
import warnings
warnings.filterwarnings('ignore')

SEED = 42
```

```python
# load the data
df_path = 'data/df_clean.csv'
df = pd.read_csv(df_path)
df = df.drop(df.columns[0], axis=1)
```

```python
# divide target variable and features
X = df.drop(['views_cat'], axis = 1)
y = df.views_cat
```

```python
# check null value
def check_missing(df):
    missing = df.isnull().sum()
    missing_percentage = (df.isnull().sum()/len(df)*100).round(2)
    missing_val = pd.concat([missing, missing_percentage], axis = 1)
    missing_val.columns = ['Missing Values', '% Missing']
    total_columns = df.shape[1]
    missing_columns = (df.isnull().sum() > 0).sum()
    print('Out of {} columns, {} columns have missing values'.format(total_columns, missing_columns))
    return missing_val

check_missing(X)
```

```python
# fillna with 0 for desc_char_length and desc_word_count
X.desc_char_length.fillna(0, inplace = True)
X.desc_word_count.fillna(0, inplace = True)
```

```python
# train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, stratify = y, random_state = SEED)
```

### Modeling

```python
# choose the optimal number of components for PCA

# Standardize the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

# plot explained variance ratio to choose the optimal number of components
pca = PCA(n_components = X_train_scaled.shape[1], random_state = SEED)
pca.fit_transform(X_train_scaled)

features = range(pca.n_components_)
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(features, np.cumsum(pca.explained_variance_ratio_))
ax.hlines(y = 0.95, xmin = 0, xmax = X_train_scaled.shape[1], color = 'r', linestyles = 'dashed')
ax.set_title('Fraction of Explained Variance', fontsize = 15)
ax.set_xlabel('Dimension #')
ax.set_ylabel('Explained Variance Ratio')
ax.set_xticks(np.arange(0, pca.n_components_ + 1, 5.0))
plt.show()
```

```python
train_score = []
cv_score = []
test_score = []
training_time = []

def build_model(input_model, model_name):

    # define which resampling method and which ML model to use in the pipeline
    scaler = StandardScaler()
    pca = PCA(n_components = 38)
    resampling = ADASYN(random_state = SEED, sampling_strategy = 1.0)
    model = input_model

    # define the pipeline, tell it to combine ADASYN with the Logistic Regression model
    pipeline = Pipeline([('Scaler', scaler), ('PCA', pca), ('ADASYN', resampling), ('Classifier', model)])
    
    # fit the data and get the train score
    start = time.time()
    pipeline.fit(X_train, y_train)
    stop = time.time()
    y_pred_train = pipeline.predict(X_train)
    train_score.append(precision_score(y_train, y_pred_train))
    
    # training time
    train_time = stop - start
    training_time.append(train_time)

    # cv score
    cv_score.append(cross_val_score(model, X_train, y_train, cv = 10, scoring = 'precision').mean())
    
    # predict
    y_pred = pipeline.predict(X_test)
    test_score.append(precision_score(y_test, y_pred))
    
    # track progress
    print('Done')

    return pipeline
```

```python
NB = build_model(GaussianNB(), 'Naive Bayes')
logreg = build_model(LogisticRegression(random_state = SEED), 'Logistic Regression')
LSVC = build_model(LinearSVC(random_state = SEED), 'Linear SVC')
sgd = build_model(SGDClassifier(random_state = SEED), 'Stochastic Gradient Descent')
KNN = build_model(KNeighborsClassifier(), 'KNN')
dt = build_model(DecisionTreeClassifier(random_state = SEED), 'Decision Tree')
rf = build_model(RandomForestClassifier(random_state = SEED), 'Random Forest')
gbm = build_model(GradientBoostingClassifier(random_state = SEED), 'Gradient Boosting Classifier')
```

```python
# take a look at the scores and go back to tune the hyperparameters if see any indication of overfitting or underfitting
classifiers = ['Naive Bayes', 'Logistic Regression', 'Linear SVM', 'Stochastic Gradient Descent', 'KNN', 'Decision Tree', 'Random Forest', 'Gradient Boosting']
result = pd.DataFrame({'Train Precision': train_score, 'CV Precision': cv_score, 'Test Precision': test_score, 'Training Time': training_time}, index = classifiers)
result.sort_values(by = 'CV Precision', ascending = False)
```

```python
# visualize the metrics
base_score = result.iloc[:, 0:3].sort_values(by = 'CV Precision', ascending = False)

fig = plt.figure()

base_score.plot.bar(rot=0, colormap = 'Set3', figsize=(20,12))
plt.title('Baseline Model Scores', fontsize = 15)

fig = plt.gcf()
fig.savefig('base_score.png', bbox_inches = 'tight')

plt.show()
```

```python
# visualize the classification report
def v_clfreport(model, title, file):
    y_pred = model.predict(X_test)
    clf_report = classification_report(y_test, y_pred, output_dict=True)
    df = pd.DataFrame(clf_report).iloc[:-1, :2].T
    
    fig, ax = plt.subplots(figsize = (6, 2))
    sns.heatmap(df, annot=True)
    plt.title(title)
    ax.set_ylim(len(df)+0, -0)
    
    fig.savefig(file, bbox_inches = 'tight')
    
    plt.show()
```

```python
# visualize classification report on the test set
v_clfreport(rf, 'Base Model Random Forest Classification Report', 'bm_rf.png')
v_clfreport(KNN, 'Base Model KNN Classification Report', 'bm_knn.png')
v_clfreport(dt, 'Base Model Decision Tree Classification Report', 'bm_dt.png')
v_clfreport(gbm, 'Base Model Gradient Boost Classification Report', 'bm_gbm.png')
v_clfreport(NB, 'Base Model Naive Bayes Classification Report', 'bm_nb.png')
v_clfreport(logreg, 'Base Model Logistic Regression Classification Report', 'bm_lr.png')
v_clfreport(sgd, 'Base Model Stochastic Graident Descent Classification Report', 'bm_sgd.png')
v_clfreport(LSVC, 'Base Model Linear SVM Classification Report', 'bm_lsvc.png')
```

### Tune Hyperparameter

```python
train_score_1 = []
cv_score_1 = []

def build_tune_model(input_model, param_dist):

    # define which resampling method and which ML model to use in the pipeline
    scaler = StandardScaler()
    pca = PCA(n_components = 38)
    resampling = ADASYN(random_state = SEED, sampling_strategy = 1.0)
    model = input_model

    # define the pipeline, tell it to combine ADASYN with the Logistic Regression model
    pipeline = Pipeline([('Scaler', scaler), ('PCA', pca), ('ADASYN', resampling), ('Classifier', model)])

    # stratified k fold
    kfold = StratifiedKFold(n_splits=3, shuffle=True, random_state=SEED)
    
    # tune hyperparameter
    pipeline = RandomizedSearchCV(pipeline, param_distributions = param_dist, scoring = 'precision', cv = kfold, n_jobs = -1, random_state = SEED)

    # fit the data
    pipeline.fit(X_train, y_train)
    
    # choose the best model
    pipeline = pipeline.best_estimator_
    
    # predict
    y_pred_train = pipeline.predict(X_train)
    train_score_1.append(precision_score(y_train, y_pred_train))
    y_pred = pipeline.predict(X_test)
    cv_score_1.append(precision_score(y_test, y_pred))
    
    # track progress
    print('Done')

    return pipeline
```

```python
# define param_dist
logreg_param = {'Classifier__C' : np.logspace(-4, 4, 20)}
LSVC_param = {'Classifier__C' : np.logspace(-4, 4, 20)}
sgd_param = {'Classifier__alpha': np.logspace(-4, 4, 20)}
knn_param = {'Classifier__n_neighbors': range(3, 10)}
dt_param = {'Classifier__max_depth': range(3, 100),
            'Classifier__min_samples_split': range(2, 40),
            'Classifier__min_samples_leaf': range(1, 5)}
rf_param = {'Classifier__n_estimators': range(50, 1000),
            'Classifier__max_depth': range(3, 100),
            'Classifier__max_features': range(3, 17),
            'Classifier__min_samples_split': range(2, 10),
            'Classifier__min_samples_leaf': range(1, 5)}
gbm_param = {'Classifier__n_estimators': range(50, 1000),
             'Classifier__learning_rate': [0.0001, 0.001, 0.01, 1, 2, 5],
             'Classifier__max_depth': range(3, 80), 
             'Classifier__min_samples_split': range(2, 10),
             'Classifier__min_samples_leaf':range(1, 5),
             'Classifier__max_features': range(3, 17)}
```

```python
logreg1 = build_tune_model(LogisticRegression(random_state = SEED), logreg_param)
LSVC1 = build_tune_model(LinearSVC(random_state = SEED), LSVC_param)
sgd1 = build_tune_model(SGDClassifier(random_state = SEED), sgd_param)
KNN1 = build_tune_model(KNeighborsClassifier(), knn_param)
dt1 = build_tune_model(DecisionTreeClassifier(random_state = SEED), dt_param)
rf1 = build_tune_model(RandomForestClassifier(random_state = SEED), rf_param)
gbm1 = build_tune_model(GradientBoostingClassifier(random_state = SEED), gbm_param)
```

```python
# take a look at the scores and go back to tune the hyperparameters if see any indication of overfitting or underfitting
classifiers = ['Logistic Regression', 'Linear SVM', 'Stochastic Gradient Descent', 'KNN', 'Decision Tree', 'Random Forest', 'Gradient Boosting']
result1 = pd.DataFrame({'Train Precision': train_score_1, 'CV Precision': cv_score_1}, index = classifiers)
result1.sort_values(by = 'CV Precision', ascending = False)
```

```python
# visualize the metrics
precision_score = result1.sort_values(by = 'CV Precision', ascending = False)

fig = plt.figure()

precision_score.plot.bar(rot=0, colormap = 'Set3', figsize=(16,8))
plt.title('Precision Scores', fontsize = 15)

fig = plt.gcf()
fig.savefig('score.png', bbox_inches = 'tight')

plt.show()
```
